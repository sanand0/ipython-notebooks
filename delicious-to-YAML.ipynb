```python
import yaml
import datetime
import lxml.html
```

```python
tree = lxml.html.parse('d:/downloads/delicious.html')

results = []
for row in tree.findall('.//a'):
    result = {
        'link': row.get('href'),
        'title': row.text.strip(),
        'date': datetime.datetime.fromtimestamp(int(row.get('add_date'))),
    }
    tag = row.get('tags', '')
    if tag:
        result['tag'] = tag
    sibling = row.getparent().getnext()
    if sibling is not None and sibling.tag == 'dd':
        result['notes'] = sibling.text.strip()
    results.append(result)
```

```python
print yaml.dump(results, default_flow_style=False, indent=2, width=80)
```
